```python
# Examples

# each chunk is a paragraph
paragraph_chunks, _, __ = pdf_to_chunks(DOCUMENT_PATH, 
                                 chunking = 'p', 
                                 grouping = 1,
                                 min_chunk_size = 100)

# each chunk is a paragraph
paragraph_pair_chunks, _, __ = pdf_to_chunks(DOCUMENT_PATH, 
                                 chunking = 'p', 
                                 grouping = 2,
                                 min_chunk_size = 100)

# each chunk is a sentence
sentence_chunks, _, __  = pdf_to_chunks(DOCUMENT_PATH, 
                                chunking = 's', 
                                grouping = 1,
                                min_chunk_size = 100)

# each chunk is 2 sentences
sentence_pair_chunks, _, __  = pdf_to_chunks(DOCUMENT_PATH, 
                                chunking = 's', 
                                grouping = 2, 
                                min_chunk_size = 100)

# each chunk is fixed size
fized_size_chunks, _, __  = pdf_to_chunks(DOCUMENT_PATH, 
                                chunking = 'f', 
                                grouping = 1, 
                                min_chunk_size = 10)
```
