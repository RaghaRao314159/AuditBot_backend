```python
from pdf2image import convert_from_path
import pytesseract
import cv2
import numpy as np

CONTENTS_PATH = "../data/documents/ar_fy2022_23_content_pages.pdf"
```

```python
image_path = "pdf_images/rgb_image.png"
rgb_images = convert_from_path(CONTENTS_PATH)
rgb_images[0].save(image_path)

rgb_image = rgb_images[0]
print(pytesseract.image_to_string(rgb_image))

```

```python
# Greyscale

prior_image = cv2.imread(image_path)
gray_image = cv2.cvtColor(prior_image, cv2.COLOR_BGR2GRAY)

preprocessed_image_path = 'pdf_images/gray_image.png'
cv2.imwrite(preprocessed_image_path, gray_image)

print(pytesseract.image_to_string(gray_image))
```

```python
# Sharpen Coloured image
kernel_1 = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

kernel_2 = np.array([[-1, -1, -1],
                   [-1,  9, -1],
                   [-1, -1, -1]])

sharpened1_rgb_image = cv2.filter2D(np.array(rgb_image), -1, kernel_1)
cv2.imwrite('pdf_images/sharpened1_rgb_image.png', sharpened1_rgb_image)

sharpened2_rgb_image = cv2.filter2D(np.array(rgb_image), -1, kernel_2)
cv2.imwrite('pdf_images/sharpened2_rgb_image.png', sharpened2_rgb_image)

print(pytesseract.image_to_string(sharpened1_rgb_image))
```

```python
print(pytesseract.image_to_string(sharpened2_rgb_image))
```

```python
# Sharpen Coloured image

sharpened1_gray_image = cv2.filter2D(np.array(gray_image), -1, kernel_1)
cv2.imwrite('pdf_images/sharpened1_gray_image.png', sharpened1_gray_image)

sharpened2_gray_image = cv2.filter2D(np.array(gray_image), -1, kernel_2)
cv2.imwrite('pdf_images/sharpened2_gray_image.png', sharpened2_gray_image)

print(pytesseract.image_to_string(sharpened1_gray_image))
```
